Cheking the oultiers in the numerical column

```python
def outliers(data,variable) :
    q3 = data[variable].quantile(0.75)
    q1 = data[variable].quantile(0.25)
    iqr = q3-q1
    lower_fence = q1-1.5*iqr
    higher_fence = q3+1.5*iqr

    return lower_fence,higher_fence
```

```python
outliers(data,'symboling')
```

Seeing the range of the outliers in the numerical columns

```python
for i in lst_num :
    print("Outliers of the {} columns is {}".format(i,outliers(data,i)))
```
